```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import re

# Setup Selenium WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Run in headless mode, comment out to see browser window
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# URL of the page containing the contract data
url = 'https://puckpedia.com/player/tyler-bertuzzi'

try:
    # Load the page
    driver.get(url)

    # Wait for the main content to load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'pp_table-wrapper')))

    # Get page source after content is dynamically loaded
    html_content = driver.page_source

    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all contract sections
    contract_sections = soup.find_all('div', class_='pp_table-wrapper')

    # List to store all contract data
    all_contracts_data = []

    for section in contract_sections:
        # Initialize a dictionary to store contract data for the current section
        contract_data = {}

        # Extract season from the section header if available
        season_header = section.find('thead').find('div').text.strip()
        contract_data['Season'] = season_header

        # Find the contract table within each section
        table = section.find('table', class_='pp_table-contract')

        if table:
            # Find all rows ('tr' elements) in the table body ('tbody')
            rows = table.find('tbody').find_all('tr')

            for row in rows:
                # Extract columns ('td' elements)
                cells = row.find_all('td')
                if len(cells) >= 2:  # Ensure it's a valid row with label and value cells
                    label = cells[0].text.strip()
                    value = cells[1].text.strip()

                    # Clean up the value if necessary (remove extra spaces, etc.)
                    value = re.sub(r'\s+', ' ', value)

                    # Store the data in the dictionary
                    contract_data[label] = value

        # Append the contract data for the current section to the list
        all_contracts_data.append(contract_data)

    # Convert the list of dictionaries to a DataFrame
    df = pd.DataFrame(all_contracts_data)

    # Display or further process the DataFrame
    print(df)

finally:
    # Close the browser
    driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the seasons and their corresponding API codes
seasons = {
    "2019-20": "YKJT4K",
    "2020-21": "CPUR97",
    "2021-22": "4ATL9N",
    "2022-23": "YPM797",
    "2023-24": "9FCTEM"
}

base_url = "https://dashboard.puckpedia.com/?q="
season_data = {}

# Function to reformat player names
def reformat_player_name(player_name):
    # Split the player name by comma
    parts = player_name.split(',')
    if len(parts) == 2:
        last_name, first_name = parts
        return f"{first_name.strip()} {last_name.strip()}"
    return player_name  # Return original if not in expected format

# Loop through each season and fetch data
for season, api_code in seasons.items():
    url = f"{base_url}{api_code}"
    
    # Step 1: Send a request to the website
    response = requests.get(url)
    
    # Step 2: Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Step 3: Extract salary information
    table = soup.find('table')  # Adjust the selector as needed
    rows = table.find_all('tr')
    
    # Extract headers
    headers = [header.text for header in rows[0].find_all('th')]
    
    # Extract data
    data = []
    for row in rows[1:]:  # Skip the header row
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]
        data.append(cols)
    
    # Step 4: Create a DataFrame
    df = pd.DataFrame(data, columns=headers)
    
    # Drop columns where all values are NaN
    df = df.dropna(axis=1, how='all')
    
    # Remove unnamed columns
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
    
    # Add a column for the season
    df['Season'] = season
    
    # Add fall_year column
    df['fall_year'] = df['Season'].str[:4].astype(int)
    
    # Reformat player names and add as a new column
    df['player_name'] = df['Playeraverages'].apply(reformat_player_name)
    
    # Drop original column if needed
    df = df.drop(columns=['Playeraverages', 'Team'])
    
    # Move 'player_name' to the first column
    columns = ['player_name'] + [col for col in df.columns if col != 'player_name']
    df = df[columns]
    
    # Store the DataFrame in the dictionary
    season_data[season] = df

# Batch rename columns in each season DataFrame
for season, df in season_data.items():
    # Remove the first 7 characters from column names starting with '20'
    new_columns = {col: col[7:] if col.startswith('20') else col for col in df.columns}
    
    # Rename columns
    df.rename(columns=new_columns, inplace=True)
    
    # Drop the first unnamed column
    if df.columns[1].startswith('Unnamed'):
        df.drop(df.columns[1], axis=1, inplace=True)
    
    # Specific renaming for the 2023-24 season
    if season == '2023-24':
        df.rename(columns={
            'Current Cap Hit': ' Cap Hit',
            'Current Salary': ' Total Salary'
        }, inplace=True)

# Convert 'Start Yr' column to year format
for season, df in season_data.items():
    if 'Start Yr' in df.columns:
        df['Start Yr'] = pd.to_datetime(df['Start Yr'], errors='coerce').dt.year

# Combine all season DataFrames into a single DataFrame
combined_df = pd.concat(season_data.values(), ignore_index=True)

```

```python
def clean_salary(sal, year):
    ''' Adds all salaries earned per year into one salary (in millions of dollars)
        "< $Minimum" is treated as $0
    '''
    if isinstance(sal, int) or isinstance(sal, float):
        return sal  # Return the salary as is if it's already numeric
    
    newlist = [0]
    for elem in sal.split('$'):
        try:
            elem = elem.replace('<','').replace(',','').replace('(TW)','').replace('Minimum','0')
            newlist.append(float(elem)*10**6)
        except ValueError:
            print(elem)
    return newlist[-1]

            
    salary = sum(newlist)/1000000
    
    return salary

# Salary Cap Data taken from spotrac.com/nba/cba
salarycap = {2019:  81.5,  2020: 81.5,   2021: 81.5,   2022: 82.5,   
             2023:   83.5,  2024: 88.0}   

def clean_and_norm_salary(sal, year):
    ''' Adds all salaries earned per year into one salary (in millions of dollars)
        "< $Minimum" is treated as $0
        Then normalizes by salary cap (in 2021 salary cap dollars)
    '''
    newlist = [0]
    for elem in sal.split('$'):
        try:
            elem = elem.replace('<','').replace(',','').replace('Minimum','0')
            elem =  float(elem)
            newlist.append(elem)
        except:
            continue
            
    salary = sum(newlist)/1000000
    salary = salary*salarycap[2021]/salarycap[year]
        
    return salary

for index, row in contract_year_fas.iterrows():
    
    prevyear = contract_year_fas.loc[index, 'prevyear']
    nextyear = contract_year_fas.loc[index, 'nextyear'] 

    prevsal = contract_year_fas.loc[index, 'previous_salary']
    nextsal = contract_year_fas.loc[index, 'next_salary']
    
    contract_year_fas.loc[index, 'PrevSalClean'] = clean_salary(prevsal, prevyear)
    contract_year_fas.loc[index, 'NextSalClean'] = clean_salary(nextsal, nextyear)
    contract_year_fas.loc[index, 'PrevSalNorm'] = clean_and_norm_salary(prevsal, prevyear)
    contract_year_fas.loc[index, 'NextSalNorm'] = clean_and_norm_salary(nextsal, nextyear)

cols = ['NextSal', 'NextSalClean','NextSalNorm']
contract_year_fas[cols].sample(8)
```
