The President's Trophy is one that is handed out to the National Hockey League's (NHL) highest point-getting team. 


Original Question: For teams that win the President's Trophy, do they have a recent history of winning the Cup/participating in the Cup Final? Is there another statistic that we should be looking at when it comes to the PT winner and their success or failure in the playoffs? 

```python
pres_trophy.head()
```

```python
#Calculating the pythagorean win percentage for the presidents trophy winning team
pres_trophy['pt_pw%'] = round(((pres_trophy['gf']**2)/((pres_trophy['gf']**2)+(pres_trophy['ga']**2)))*100)
```

```python
#Calculating the pythagorean win percentage for the Cup winning team
pres_trophy['cw_pw%'] = round(((pres_trophy['cw_gf']**2)/((pres_trophy['cw_gf']**2)+(pres_trophy['cw_ga']**2)))*100)
```

```python
#How did the Cup winner stack up against the PT winner in terms of GA and GF? What was the difference between the two teams when it came to the pw%?
pres_trophy['gf_champ_vs_pt'] = pres_trophy['cw_gf']-pres_trophy['gf']
pres_trophy['ga_champ_vs_pt'] = pres_trophy['cw_ga']-pres_trophy['ga']
pres_trophy['pw%_champ_vs_pt'] = pres_trophy['cw_pw%']-pres_trophy['pt_pw%']
```

33 times the presidents trophy winner did not win the stanley cup. But why? What was it about the team that opened them up to this shortfall? Success can be measured in more than one way in hockey. Points isn't necessarily the best measure - two teams can have the same number of points, but not the same number of wins. And what about goal scoring or preventing ability? I've always said "Defense and goaltending win cups". My dad said it, his dad said it, and that's why I say it. It feels right. This is the first time that I was able to actually see how that is true. First we'll look at the non-Cup winners to see how that pans out and then we will see if it holds up with the Cup winners.

A negative number in the 'ga_champ_v_pt' indicates that the Cup winning team had a better goals against statistic compared to the team that won the Presidents Trophy. That championship team had allowed X number of goals _fewer_ than the top point-getting team, thus they were a much more defensive team compared to the PT winner.

For example, in 2022, the Florida Panthers finished the season with the most number of points in the entire league. Yet they lost in the 2nd Round of the playoffs. The Stanley Cup winning Colorado Avalanche, in comparison, allowed 12 fewer goals in the season compared to the Panthers. 

This isn't the case for _every_ team that wins the Cup that didn't win the president's trophy, but this phenomenon happens quite often in this data set. More often than instances where the winning team had a higher goals for statistic compared to the PT team. 

Out of the 33 non-Cup PT winners, 13 times we see the eventual Champion have a _lower_ goals against statistic.

In the case of the PT winners who went on to be Cup winners, in nearly every instance they were the better team both offensively and defensively. They were a "complete" team. I would even make the argument that you could, and should, consider the instances when the winners do not follow that trend as outliers because of the era in which they existed.

The 2001 Stanley Cup Final took place during a time known as the "Dead Puck Era" where scoring was incredibly low and playing the trap was king. And any Final before 1991 was before the "Modern Era", where the game was different.

You'll notice I had purposely excluded the year 2019 from the dataframe; I included the 2019 Cup Final in this original dataframe specifically to address what happened to the Bruins.

When it comes down to it, the teams that win the PT are not losing to teams that are horribly worse than they are. 

People like to bring up the 2019 Tampa Bay Lightning as the perfect example of a team that was "meant" to win it all go and lay a giant egg. In fact, this was the team that made me wonder how much playoff experience a team has had impacts their winning of the Presidents Trophy and a Cup. Turns out, teams that are good, end up with presidents trophies mixed in with Stanley Cups. Sometimes the PT win comes at the start or early on in the beginning of a dynasty (2019 TBL).
```python
advanced_stats['reg_ga'] = advanced_stats['GA'] - (advanced_stats['PPA'] + advanced_stats['SHA'])
advanced_stats['reg_gf'] =  advanced_stats['GF'] - (advanced_stats['PP'] + advanced_stats['SH'])
advanced_stats.head()
```

```python
# goals for and against diff with JUST power play goals and with non-power play goals

#pwp based on PP goals
advanced_stats['pp_pw%'] = round(((advanced_stats['PP']**2)/((advanced_stats['PP']**2)+(advanced_stats['PPA']**2)))*100)

#pwp based on regular only goals
advanced_stats['reg_pw%'] = round(((advanced_stats['reg_gf']**2)/((advanced_stats['reg_gf']**2)
                                                                  +(advanced_stats['reg_ga']**2)))*100)

#pwp based on sh only goals
advanced_stats['sh_pw%'] = round(((advanced_stats['SH']**2)/((advanced_stats['SH']**2)
                                                                  +(advanced_stats['SHA']**2)))*100)

advanced_stats.head()
```

```python
advanced_stats.columns
```

```python
#calculating the percentage points difference between the president's trophy winner and all of the other teams 

def diff_advanced(row):
    index = advanced_stats.index.get_loc(row.name)
    if index == 0:
        return row
    
    pres_winner = advanced_stats.iloc[0]
    
    row['pp_pw_diff'] = row['pp_pw%'] - pres_winner['pp_pw%']
    row['ppa_diff'] = row['PPA'] - pres_winner['PPA']
    row['pp_diff'] = row['PP'] - pres_winner['PP']
    row['sh_pw_diff'] = row['sh_pw%'] - pres_winner['sh_pw%']
    row['sha_diff'] = row['SHA'] - pres_winner['SHA']
    row['sh_diff'] = row['SH'] - pres_winner['SH']
    row['reg_pw_diff'] = row['reg_pw%'] - pres_winner['reg_pw%']
    row['reg_ga_diff'] = row['reg_ga'] - pres_winner['reg_ga']
    row['reg_gf_diff'] = row['reg_gf'] - pres_winner['reg_gf']
    row['gf_diff'] =  row['GF'] - pres_winner['GF']
    row['ga_diff'] = row['GA'] - pres_winner['GA']
    row['sh_diff'] = row['PP'] - pres_winner['PP']
    row['reg_pw_diff'] = row['reg_pw%'] - pres_winner['reg_pw%']
    row['reg_gf_diff'] = row['reg_gf'] - pres_winner['reg_gf']
    row['reg_ga_diff'] = row['reg_ga'] - pres_winner['reg_ga']

    return row

# Apply the function to create new columns
advanced_stats = advanced_stats.apply(diff_advanced, axis=1)
column_order = ['Team', 'PTS', 'GF', 'GA', 'pw%', 'pp_pw%', 'PPA', 'PP','sh_pw%','SH', 'SHA', 'reg_pw%', 'reg_ga', 'reg_gf', 'pp_pw_diff','ppa_diff', 'pp_diff',
             'sh_pw_diff', 'sha_diff', 'sh_diff', 'reg_pw_diff',  'reg_ga_diff', 'reg_gf_diff', 'gf_diff', 'ga_diff',  'reg_pw_diff',  'reg_ga_diff','reg_gf_diff' ]
advanced_stats = advanced_stats[column_order]

advanced_stats.head()
```

```python
eastern_conference['pwinp'] = (eastern_conference['GF']**2)/((eastern_conference['GF']**2)+(eastern_conference['GA']**2))
eastern_conference['winp'] = eastern_conference['win']/eastern_conference['GP']
eastern_conference['gf%'] = eastern_conference['GF']/(eastern_conference['GF']+eastern_conference['GA'])
eastern_conference['ga%'] = eastern_conference['GA']/(eastern_conference['GF']+eastern_conference['GA'])
eastern_conference['goal_diff'] = eastern_conference['GF']-eastern_conference['GA']
```

```python
def diff_ecf(row):
    index = eastern_conference.index.get_loc(row.name)
    if index == 0:
        return row
    
    conf_champs = eastern_conference.iloc[0]
    
    row['Diff_GF'] = row['gf%'] - conf_champs['gf%']
    row['Diff_GA'] = row['ga%'] - conf_champs['ga%']
    row['diff_goal_dfs'] = row['goal_diff'] - conf_champs['goal_diff']
    return row

# Apply the function to create new columns
eastern_conference = eastern_conference.apply(diff_ecf, axis=1)
eastern_conference
```
